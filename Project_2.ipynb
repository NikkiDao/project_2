```python
import pandas as pd
from sqlalchemy import create_engine
from key import keys
```

## EXTRACT- Big City Health

```python
#Read the big city health csv file
big_city = "/Users/anthonygarcia/Desktop/project2_nonrepo/Big_Cities_Health_Data_Inventory.csv"
city_df = pd.read_csv(big_city)
city_df.head()
```

## TRANSFORM - Big City Health

```python
#split city and state into two columns
city_df['City'] = city_df['Place'].str.split(',').apply(lambda x: x[0])
city_df['State'] = city_df['Place'].str.split(',').apply(lambda x: x[-1])
city_df
```

```python
#change the city names to exclude county name 
city_df['City']=city_df['City'].replace(["Atlanta (Fulton County)", "Miami (Miami-Dade County)","San Diego County","Las Vegas (Clark County)","Fort Worth (Tarrant County)","Portland (Multnomah County)"], 
                            ["Atlanta","Miami","San Diego","Las Vegas","Fort Worth","Portland"])
```

```python
#view city column
city_df.City.value_counts()
```

```python
#Create a new table with the selected columns
new_city_df = city_df[['Indicator','Value','City','State']].copy()
new_city_df.head()
```

```python
#Reorder columns
df = new_city_df[['City','State','Indicator','Value']]
df
```

```python
#Create a list of values used for data
types = ['Diabetes Mortality Rate (Age-Adjusted; Per 100,000 people)',
         'Heart Disease Mortality Rate (Age-Adjusted; Per 100,000 people)',
        'Percent of Adults Who Are Obese',
        'Percent of High School Students Who Are Obese']
```

```python
#Filter table by using the created list
#Final Table
filtered= df[df['Indicator'].isin(types)]
filtered
```

```python
#Group city and indicator to get average 
health_data = filtered.groupby(['City', 'Indicator']).mean().round(2).reset_index()
health_data
```

```python
#rename columns
health_df =health_data.rename(columns = {'City':'city', 'Indicator':'medical_cause', 'Value':'measure'})
health_df
```

## LOAD - Big City Health

```python
#connect to local database
protocol = 'postgresql'
username = 'postgres'
password = keys
host = 'localhost'
port = 5432
database_name = 'health_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string) 
```

```python
#import city health data into sql table
health_df.to_sql(name='big_city', con=engine, if_exists='append', index=False) 
```

```python
#check and see that data loaded
pd.read_sql_query('select * from big_city', con=engine).head()
```

## EXTRACT - Fast Food Restaurant

```python
#Read the fast food restaurant csv file
fast_food = "/Users/anthonygarcia/Desktop/project2_nonrepo/FastFoodRestaurants.csv"
food_df = pd.read_csv(fast_food)
food_df.head()
```

## TRANSFORM - Fast Food Restaurant

```python
#Create a new table with the selected columns
food_ = food_df[['city','name']].copy()
food_
```

```python
# combine restaurants with cities to see total amount of restaurants per city
restaurants =food_.groupby(['city']).count()
restaurants
```

```python
#create new dataframe from groupby
food_data = pd.DataFrame(restaurants)
food_data
```

```python
#rename column
rename =food_data.rename(columns = {'name':'restaurant_count'})
rename
```

```python
#reset the index
food_data= rename.reset_index()
food_data
```

## LOAD - Fast Food Restaurant

```python
#import fast food data into sql table
food_data.to_sql(name='fast_food', con=engine, if_exists='append', index=False)
```
